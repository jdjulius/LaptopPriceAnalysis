```python
var_num = pricing_data.select_dtypes(include=['float64', 'int']).columns
var_cat = pricing_data.select_dtypes(include=['object', 'category']).columns

print(var_num)
print(var_cat)
```

```python
for i, colum in enumerate(var_num):
    fig, ax = plt.subplots(1, 3, figsize=(15, 3.84))
    fig.suptitle('Grficos: variables cuantitativas ' + colum)

    sns.histplot(x=pricing_data[colum], data=pricing_data, kde= True, ax=ax[0])
    ax[0].axvline(utils.median(pricing_data[colum]), color='red', linestyle='dashed', linewidth=2,label=f'Media: {utils.median(pricing_data[colum]):.1f}')
    ax[0].axvline(utils.mean(pricing_data[colum]), color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {utils.mean(pricing_data[colum]):.1f}')

    sns.boxplot(x=pricing_data[colum], ax=ax[1])
    ax[1].axvline(utils.median(pricing_data[colum]), color='red', linestyle='dashed', linewidth=2,label=f'Media: {utils.median(pricing_data[colum]):.1f}')
    ax[1].axvline(utils.mean(pricing_data[colum]), color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {utils.mean(pricing_data[colum]):.1f}')

    sns.violinplot(data=pricing_data, x=colum, showfliers=False, ax=ax[2])
    ax[2].axvline(utils.median(pricing_data[colum]), color='red', linestyle='dashed', linewidth=2,label=f'Media: {utils.median(pricing_data[colum]):.1f}')
    ax[2].axvline(utils.mean(pricing_data[colum]), color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {utils.mean(pricing_data[colum]):.1f}')
```

```python
plt.figure(figsize=(10, 8))
corr_matrix = pricing_data[var_num].corr()
sns.heatmap(corr_matrix, annot=True, cmap='cubehelix', vmin=-1, vmax=1)
plt.show()
```
